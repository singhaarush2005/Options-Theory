```python
def annual_vol(df):
    """
    Function to calculate annual volatility

    Parameters
    ----------
    df : dataframe
         Stock info

    Returns
    -------
    vol : float
        annual volatility of the stock
    """

    log_return = np.log(df.Close/df.Open)
    log_return = log_return.dropna()
    daily_vol = log_return.std()
    vol = daily_vol*np.sqrt(252)


    return vol
```
